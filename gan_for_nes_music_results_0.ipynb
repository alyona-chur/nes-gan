```python
import librosa
import IPython.display as ipd
import os

paths = {}
for file in os.listdir('generated_wav'):
  file_path = os.path.join('generated_wav', file)
  if os.path.splitext(file_path)[1] != '.wav':
    continue
  paths[file] = file_path
```

```python
# Interation 0
# Before training

data, _ = librosa.load(paths['run-12-iteration-0-batch-size-512_0.wav'])
ipd.Audio(data, rate=22050)
```

```python
# Iteration 1000
# Some notes and noices start to appear

data, _ = librosa.load(paths['run-12-iteration-1000-batch-size-512_0.wav'])
ipd.Audio(data, rate=22050)
```

```python
# Iteration 5000
# Melody flow makes more sense (still sounds like " ")

data, _ = librosa.load(paths['run-12-iteration-5000-batch-size-512_0.wav'])
ipd.Audio(data, rate=22050)
```

```python
# Iteration 9000
# More different instruments (there are 4 instruments in NES) are noticable

data, _ = librosa.load(paths['run-12-iteration-9000-batch-size-512_0.wav'])
ipd.Audio(data, rate=22050)
```

```python
# Iteration 10000 - one of the best
# Melody is not homogeneous anymore

data, _ = librosa.load(paths['run-12-iteration-10000-batch-size-512_1.wav'])
ipd.Audio(data, rate=22050)
```

```python
# Iteration 10000 - one of the best

data, _ = librosa.load(paths['run-12-iteration-10000-batch-size-512_4.wav'])
ipd.Audio(data, rate=22050)
```
